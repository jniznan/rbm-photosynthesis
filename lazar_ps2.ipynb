```python
import matplotlib.pyplot as plt
from pysb import bng
import pysb
from tools import dfs, rules, context_elimination, bng_simulate, reaction_network, to_bngl
import pandas as pd
import matplotlib
import numpy as np
import networkx as nx
%matplotlib inline
```

```python
lazar_sim = pd.read_csv('models/lazar/original/sim.csv')
```

```python
with open('models/lazar_ps2/bngl/lazar_model8.bngl') as f:
    model = ''.join(f.readlines())
```

```python
res = bng_simulate(model, lazar_sim['time'])[1:]
```

```python
time = res['time']
```

```python
def Funq(res):
    Qa_m = res['Qa_m']
    return 0.45 * Qa_m / (1 - 0.55 * Qa_m)

def Fq(res):
    Qa_m, PQ_n = res['Qa_m'], res['PQ']
    return 0.45 * Qa_m / (1 - 0.55 * Qa_m) / (1 + (1./45 + 4./63 * Qa_m) * PQ_n)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(time, Funq(res), label='fluorescence - unquenched', linewidth=2)
plt.plot(time, Fq(res), label='fluorescence - quenched', linewidth=2)
plt.legend(loc='upper left')
plt.xscale('log')
plt.xlabel('time')
#plt.savefig('lazar_ps2_sim.pdf', bbox_inches='tight')
```
