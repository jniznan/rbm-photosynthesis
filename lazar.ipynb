```python
from models.lazar.naive import model
import matplotlib.pyplot as plt
from pysb import bng
import pysb
from tools import dfs, rules, context_elimination, bng_simulate, reaction_network, to_bngl
import pandas as pd
import matplotlib
import numpy as np
import networkx as nx
%matplotlib inline
```

Load the results of simulation from e-photosynthesis.org

```python
lazar_sim = pd.read_csv('models/lazar/original/sim.csv')
```

Simulate the rewritten model in bngl

```python
res = bng_simulate(model, lazar_sim['time'])[1:]
```

```python
time, Qa_m, PQ_n, Pc_p, P700_p = res['time'], res['Qa_m'], res['PQ_n'], res['Pc_p'], res['P700_p']
```

Compute fluorescence and I820

```python
def Fq(res):
    Qa_m, PQ_n = res['Qa_m'], res['PQ_n']
    return 0.45 * Qa_m / (1 - 0.55 * Qa_m) / (1 + (1./45 + 4./63 * Qa_m) * PQ_n)

def Funq(res):
    Qa_m = res['Qa_m']
    return 0.45 * Qa_m / (1 - 0.55 * Qa_m)

def I820(res):
    Pc_p, P700_p = res['Pc_p'], res['P700_p']
    return 10 ** ( -2.16e-7 * (1590 * Pc_p + 10300 * P700_p))
```

Visual comparison of the results with the model on e-photosynthesis.org

```python
plt.plot(time, Fq(res))
plt.plot(time, lazar_sim['Fq(t)'])
plt.xscale('log')
```

```python
plt.plot(time, Funq(res))
plt.plot(time, lazar_sim['Funq(t)'])
plt.xscale('log')
```

```python
plt.plot(time, I820(res))
plt.plot(time, lazar_sim['I820'])
plt.xscale('log')
```

#### Stochastic

Must convert deterministic rates to stochastic ones:

```python
NA = 6.02e23
V = 1e-20
c = NA * V
c
```

```python
# Rescale initial counts:
for p in m2.parameters:
    if p.name[:4] == 'init':
        p.value *= c
```

```python
# Rescale the rates of second order equations:
second_order_rates = set()
for r in m2.rules:
    if len(r.reactant_pattern.complex_patterns) == 2:
        second_order_rates.add(r.rate_forward)
        if r.rate_reverse:
            second_order_rates.add(r.rate_reverse)
for p in second_order_rates:
    p.value /= c
```

```python
# Perform 20 stochastic simulations
sim = []
for i in range(20):
    sim.append(bng_simulate(m2, time, 'ssa')[1:])
```

```python
# Scale observables back to the same scale as in the deterministic simulation:
for r in sim:
    for name in r.dtype.names:
        if name != 'time':
            r[name] /= c
```

```python
matplotlib.rc('font', size=12)
plt.figure(figsize=(17, 5))
plt.subplot(121)
plt.plot(time, map(np.mean, zip(*map(I820, sim))), label='$I_{820}$ - SSA', linewidth=2)
plt.legend(loc='lower left')
plt.xscale('log')
plt.xlabel('time')
plt.subplot(122)
plt.plot(time, map(np.mean, zip(*map(Fq, sim))), label='$F_q$ - SSA', linewidth=2)
plt.plot(time, map(np.mean, zip(*map(Funq, sim))), label='$F_{unq}$ - SSA', linewidth=2)
plt.legend(loc='upper left')
plt.xscale('log')
plt.xlabel('time')
#plt.savefig('lazar_rb_sim_ssa.pdf', bbox_inches='tight')
```
