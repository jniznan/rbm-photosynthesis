```python
from models.lazar.reduced import model
import matplotlib.pyplot as plt
from pysb import bng
import pysb
from tools import dfs, rules, context_elimination, bng_simulate, reaction_network, to_bngl
import pandas as pd
import matplotlib
import numpy as np
import networkx as nx
%matplotlib inline
```

```python
lazar_sim = pd.read_csv('models/lazar/original/sim.csv')
```

```python
res = bng_simulate(model, lazar_sim['time'])[1:]
```

```python
time, Qa_m, PQ_n, Pc_p, P700_p = res['time'], res['Qa_m'], res['PQ_n'], res['Pc_p'], res['P700_p']
```

```python
Fq = 0.45 * Qa_m / (1 - 0.55 * Qa_m) / (1 + (1./45 + 4./63 * Qa_m) * PQ_n)
Funq = 0.45 * Qa_m / (1 - 0.55 * Qa_m)
i820 = 10 ** ( -2.16e-7 * (1590 * Pc_p + 10300 * P700_p))
```

```python
with open('models/replace_ps2/bngl/model.bngl') as f:
    exchange_model = '\n'.join(f.readlines())
```

```python
res = bng_simulate(exchange_model, lazar_sim['time'])[1:]
```

```python
time, Qa_m, PQ_n, Pc_p, P700_p = res['time'], res['Qa_m'], res['PQ_n'], res['Pc_p'], res['P700_p']
```

```python
Fq1 = 0.45 * Qa_m / (1 - 0.55 * Qa_m) / (1 + (1./45 + 4./63 * Qa_m) * PQ_n)
Funq1 = 0.45 * Qa_m / (1 - 0.55 * Qa_m)
i8201 = 10 ** ( -2.16e-7 * (1590 * Pc_p + 10300 * P700_p))
```

```python
matplotlib.rc('font', size=12)
plt.figure(figsize=(17, 5))
plt.subplot(121)
plt.plot(time, Fq, label='original model', linewidth=2)
plt.plot(time, Fq1, label='new model', linewidth=2)
plt.xscale('log')
plt.xlabel('time')
plt.legend(loc='upper left')
plt.title('Fluorescence - quenched')
plt.ylim((0, 1.05))
plt.subplot(122)
plt.plot(time, Funq, label='original model', linewidth=2)
plt.plot(time, Funq1, label='new model', linewidth=2)
plt.legend(loc='upper left')
plt.title('Fluorescence - unquenched')
plt.xscale('log')
plt.xlabel('time')
plt.ylim((0, 1.05))
#plt.savefig('lazar_ex_flu.pdf', bbox_inches='tight')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(time, i820, label='original model', linewidth=2)
plt.plot(time, i8201, label='new model', linewidth=2)
plt.legend(loc='lower left')
plt.title('$I_{820}$')
plt.xscale('log')
plt.xlabel('time')
plt.ylim((0.997, 1.0001))
#plt.savefig('lazar_ex_i820.pdf', bbox_inches='tight')
```
