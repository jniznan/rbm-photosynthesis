```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
%matplotlib inline
```

## Lazar's model

```python
lazar_sim = pd.read_csv('models/lazar/original/sim.csv')
```

```python
time = lazar_sim['time']
i820 = lazar_sim['I820']
fluo_q = lazar_sim['Fq(t)']
fluo_unq = lazar_sim['Funq(t)']
```

```python
matplotlib.rc('font', size=12)
plt.figure(figsize=(17, 5))
plt.subplot(121)
plt.plot(time, i820, label='I820', linewidth=2)
plt.legend()
plt.xscale('log')
plt.xlabel('time')
plt.subplot(122)
plt.plot(time, fluo_q, label='fluorescence - quenched', linewidth=2)
plt.plot(time, fluo_unq, label='fluorescence - unquenched', linewidth=2)
plt.legend(loc='upper left')
plt.xscale('log')
plt.xlabel('time')
#plt.savefig('lazar_sim.pdf', bbox_inches='tight')
```
